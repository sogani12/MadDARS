```python
major = []
for val in d["h1"]:
    major.append(val)

for h in [item for item in d.keys() if item.startswith("h2")]:
    num = d[h][80]
    if len(d[h][2]) >= 1:
        major.append(tuple(d[h][2]))
    else:
        l = list(combinations(d[h][1],num))
#         key = l[0]
        val = tuple(sorted(l))
        major.append(val)
        
        
major
```

```python
def grab_key(data):
    if isinstance(data, str):
        return int(re.findall(r"\d+",data)[0])
    return grab_key(data[0])
```

```python
major = sorted(major, key=lambda item: grab_key(item))
```

```python
major
```

```python
refl = []
subl = []
courses = set()
ref_idx = 0

for course in major:
    if isinstance(course, str):
        courses.add(course)
        refl.append(course)
        ref_idx += 1
    else:
        sub = []
        add_sub = False
        for path in course:
            if not add_sub:
                hits = 0
                for c in path:
                    if c in courses:
                        hits += 1
                if hits == 0:
                    refl.append(path)
                    courses.update(path)
                    add_sub = True
                    ref_idx+=1
                    length = len(path)
                else:
                    sub.append(path)
            else:
                sub.append(path)
        subl.append((ref_idx+1, length, tuple(sub)))

refl, subl
```

```python
def clean_key(string):
    i = re.findall(r"(\D+)(\d+)", string)[0]
    key = sorted(i[0].split("/"))[0]
    code = i[1]
    return key+code
```

```python
clean_key('COMPSCI/MATH240')
```

```python
ref = []
for r in refl:
    if isinstance(r, str):
        ref.append(clean_key(r))
    else:
        for i in r:
            ref.append(clean_key(i))
            
ref = ",".join(ref)
ref
```

```python
variants = []

for pos, length, sub in subl:
    for s in sub:
        variants.append((pos, length, ",".join(s)))
        
# variants
```
